### Import Statement

```python
import pandas as pd
import re
import numpy as np
import seaborn as sns
```

### The Dataframe

```python
df = pd.read_csv("WELLCOME_APCspend2013_forThinkful.csv")
df
```

```python
df = df.sort_values(['Publisher', 'Journal title'], axis=0, ascending=True, 
                    inplace=False, kind='quicksort', na_position='last')
df
```

```python
# Cleaning up the 'Publisher' column.
df['Publisher'] = df['Publisher'].str.replace('ACS', 'ACS (American Chemical Society) Publications'
                    ).str.replace('AMERICAN CHEMICAL SOCIETY', 'ACS (American Chemical Society) Publications'
                    ).str.replace('Wiley-Blackwell, John Wiley & Sons', 'Wiley-Blackwell'
                    ).str.replace('Wiley/Blackwell', 'Wiley-Blackwell')
df
```

```python
# Standardizing the capitalization in the 'Article title' column.
df['Article title'] = df['Article title'].str.capitalize()
df
```

```python
# Stripping whitespace from the 'Journal title' column.
df['Journal title'] = df['Journal title'].str.strip()
df
```

```python
# Dropping missing values from the 'Journal title' column.
df['Journal title'] = df['Journal title'].dropna()
df
```

```python
df['Journal title'].value_counts()
```

```python
# After running .count() on the 'Journal title' column, I realized it needed to be cleaned up a bit. 
df['Journal title'] = df['Journal title'].str.replace('PLoS One', 'PLos One').str.replace('PLoS ONE', 'PLos One')
df
```

```python
df['Journal title'].value_counts()
```

```python
# I need to find statistics on the costs now and because the cost column name is lengthy, I'm going to shorten it.
df.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)':'Cost'}, inplace=True)
df
```

```python
# Removing the '£' sign from the 'COST...' column.
df['Cost'] = df['Cost'].str.replace('£', '')
df
```

```python
# Removing the '$' sign from the 'COST...' column.
df['Cost'] = df['Cost'].str.replace('$', '')
df
```

```python
# Removing outliers from the 'Cost' column.
df['Cost'] = df['Cost'].str.replace('999999', '').str.replace('201024', '').str.replace('192645', ''
                ).str.replace('13200', '')
```

```python
df['Cost'] = df['Cost'].astype(float)
df
```
