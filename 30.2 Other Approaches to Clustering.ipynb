### Appendix

Explanation of the cluster types:

- K-means: an interative algorithm that eventually converges on a solution.
- Mean Shift: calculates the probability that a data point will be present at any point in the n-dimensional space designed by the number of features.
- Spectral Clustering: based on quantifying similarities between data points.
- Affinity Propagation: based on defining examples for data poitns.
