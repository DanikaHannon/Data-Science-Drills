__One Variable Plotted Four Different Ways__

_1. Scatter Plot_

```python
flights = sns.load_dataset('flights')
sns.relplot(x='year', y='passengers', data=flights);
print('Pros: this graph clearly shows the number of passengers on flights increases each year.')
print('Cons: having multiple data points on the same x-coordinate is confusing.')
```

_2. Line Plot with 95% Confidence Interval_

```python
sns.relplot(x='year', y='passengers', kind='line', 
            data=flights)
print('Pros: it is easy to see the number of passengers increasing year over year.')
print('Cons: the confidence interval is not very useful with this data')
```

_3. Line Plot without 95% Confidence Interval_

```python
sns.relplot(x='year',  y='passengers', kind='line', 
            ci=None, data=flights)
print('Pros: it is easy to see the number of passengers increasing year over year.')
print('Cons: the grid background is distracting and would be better if the white lines were replaced with black lines and the grey changed to white.')
```

_4. Line Plot without the Aggregator Function_

```python
sns.relplot(x='year',  y='passengers', estimator=None, 
            kind='line', data=flights)
print('Pros: like the other graphs, it is easy to see that the number of passengers increases over time.')
print('Cons: the spikes are confusing and are not a good representation of the data.')
```

__Choose two continuous variables and plot them three different ways.__

#### The Data Set

```python
df = sns.load_dataset('fmri')
df.head()
```

_1. A Lineplot with Aggregated Data_

```python
sns.relplot(x='timepoint',  y='signal', kind='line', 
            estimator=None, data=df)
print('Pros: the graph shows the connection between the timepoint and line.')
print('Cons: the data does not look smooth - and I saw an example where it looked more even. In the next graph, I will run the example code.')
```

_2. A Lineplot without Aggregated Data_

```python
sns.relplot(x='timepoint', y='signal', hue='subject',
            col='region', row='event', height=3,
            kind='line', estimator=None, data=df)
print('Pros: because the subjects are shown individually, it is easier to see the variety in their data.')
print('Cons: the graphs are busy and crowded. It could be helpful to look at a few subjects at a time instead of all the subjecst at once.')
```

_3. A Boxplot_

```python
sns.boxplot(x='timepoint', y='signal', hue='subject',
           data=df)
sns.despine(offset=20, trim=True)
print('Pros: the timepoints are spread out nicely.')
print('Cons: this data does not seem well-suited for a box plot. And I am having trouble getting the legend moved to the right place.')
```

__Choose one continuous variable and one categorical variable, and plot them six different ways.__

_1. A Scatterplot_

```python
sns.catplot(x='origin', y='acceleration', data=df)
print('Pros: it is easy to see the range in acceleration based on the country of origin.')
print('Cons: in some places, the datapoints are densley clustered together, which makes the graph harder to read.')
```

_2. A Scatterplot iwth the "Swarm" Feature_

```python
sns.catplot(x='origin', y='acceleration', kind='swarm', data=df)
print('Pros: with the "swarm" feature added, it is easier to understand the density of the data points.')
print('Cons: the data is not arranged from largest to smallest and would be easier to read if the countries of origin were moved around.')
```
